### Проверка правильности расчета эффективности обогащения  
Проверили расхождение для обогащения расчитанного в данных и расчитанного рассчитанного вручную:

```python
C = df_train['rougher.output.concentrate_au']
F = df_train['rougher.input.feed_au']
T = df_train['rougher.output.tail_au']

df_train['rougher.output.recovery_handcalc'] = (C * (F - T)) / (F * (C - T)) * 100

MAE = mean_absolute_error(df_train['rougher.output.recovery_handcalc'],
                    df_train['rougher.output.recovery'])
print(f'Расхождения в расчетах между предоставленными данными и расчитанными нами: {MAE}')

# Удалили расчитаное вручную значение обогащения
df_train = df_train.drop('rougher.output.recovery_handcalc',axis=True)
```

**Вывод:**
- Расхождения в расчетах очень мало, значит **предоставленные данные верны**.

```python
# Заполнили пропуски в тестовой выборке
df_test = df_test.ffill(axis=0)

# Добавили целевые признаки из полной выборки
df_test['rougher.output.recovery'] = df_full.loc[df_test.index,'rougher.output.recovery']
df_test['final.output.recovery'] = df_full.loc[df_test.index,'final.output.recovery']

# Заполнили пропуски в полной выборке, предварительно изъяв целевые признаки
target_rougher = df_full['rougher.output.recovery']
target_final = df_full['final.output.recovery']
features = df_full.drop(['final.output.recovery','rougher.output.recovery'], axis=1)
features.ffill(axis=0,inplace=True)

# Добавили целевые признаки обратно и удалили строки с пропусками в них
features['rougher.output.recovery'] = target_rougher
features['final.output.recovery'] = target_final
df_full = features.dropna()

# Повторили все действия на обучающей выборке
target_rougher = df_train['rougher.output.recovery']
target_final = df_train['final.output.recovery']
features = df_train.drop(['final.output.recovery','rougher.output.recovery'], axis=1)
features.ffill(axis=0,inplace=True)
features['rougher.output.recovery'] = target_rougher
features['final.output.recovery'] = target_final
df_train = features.dropna()
```

## Модель

### Функция для вычисления итоговой sMAPE
Написали функцию для расчета sMAPE и финального sMAPE

```python
def sMAPE_corr(target, prediction):
    value = (np.abs(target - prediction) / ((abs(target) + abs(prediction)) / 2)).mean() * 100
    return value
```

```python
def fin_sMAPE(rougher_sMAPE, final_sMAPE):
    value = 0.25 * rougher_sMAPE + 0.75 * final_sMAPE
    return value
```

```python
linear_model = LinearRegression()
linear_rougher_score = cross_val_score(linear_model, X_train, rougher_y_train, cv=5, scoring=scorer).mean()
linear_final_score = cross_val_score(linear_model, X_train, final_y_train, cv=5, scoring=scorer).mean()
fin_linear_score = fin_sMAPE(linear_rougher_score, linear_final_score)
print('sMAPE для первого целевого признака:', -linear_rougher_score)
print('sMAPE для второго целевого признака:', -linear_final_score)
print('Итоговый показатель sMAPE:', -fin_linear_score)
```

#### Случайный лес
Обучили модель Случаного леса и проверили её эффективность:


```python
best_depth = 0
best_est = 0
best_fin_sMAPE = 100
best_forest_rougher_score = 100
best_forest_final_score = 100

for depth in range(3,8):
    for est in range(1,26,5):
        forest_model = RandomForestRegressor(random_state=52, n_estimators=est, max_depth=depth)
        forest_rougher_score = cross_val_score(forest_model, X_train, rougher_y_train, cv=5, scoring=scorer).mean()
        forest_final_score = cross_val_score(forest_model, X_train, final_y_train, cv=5, scoring=scorer).mean()
        fin_forest_score = fin_sMAPE(forest_rougher_score, forest_final_score)
        print(depth, est, fin_forest_score)
        if -fin_forest_score < best_fin_sMAPE:
            best_fin_sMAPE = - fin_forest_score
            best_depth = depth
            best_est = est
            best_forest_rougher_score = -forest_rougher_score
            best_forest_final_score = -forest_final_score
print('sMAPE для первого целевого признака:', best_forest_rougher_score)
print('sMAPE для второго целевого признака:', best_forest_final_score)
print(f'Лучший результат итогового значения sMAPE: {best_fin_sMAPE}\n\
        При глубине {best_depth}, и количестве деревьев {best_est}')
```
